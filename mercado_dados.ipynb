### Fontes de dados 
Por que coletamos dados?
- Avaliar desempenho
- Identificar as dores de um negócio. Onde estão o gargalos, os famosos gaps?
- Criar um novo produto

Nós coletamos dados para **criar estratégias**

Para coletar dados, precisamos estabelecer 3 etapas: metas, indicadores de desempenho e métricas

- METAS
É o que queremos alcançat: objetivo + valor+ prazo
exemplo: aumentar o faturamento em 10% em 12 meses

- INDICADORES DE DESEMPENHO
KPIs são o monitoramento do "como" definimos a estratégias para chegar na meta.
Exemplo: novos clientes, taxa de satisfação

- MÉTRICAS
São resultados que irão abastecer os indicadores. Podem ser qualitativas com depoimentos de clientes. 
Exemplo: Quantidade de Clientes, Compras, etc...


**DADOS PRIMÁRIOS E DADOS SECUNDÁRIOS**
DADOS PRIMÁRIOS: são dados coletados de forma direta. Como transação, produtos e funcionários

O dados que eu coleto na Rede são dados primários.

DADOS SECUNDÁRIOS: dado coletado de forma indireta. Como: Dados de Mercado, Google Analytics, Open Data. Usando novamente a Rede como exemplo, a Rede faz uso de dado secundário seja de boletins, Anuários, IPEA, IBGE, SSP etc...

**TIPOS DE DADOS**
- Dados estruturados
Basicamente são dados armazenados e acessados com uma maior facilidade, porque eles possuem formato fixo. Ou seja, se eu tenho um banco de dados em SQL ou uma planilha em excel ou google sheets, toda a estrutura de coluna com uma chave e um valor ela vai seguir um padrão, já tem um comportamento pré-determinado para conseguir fazer uma análise desses dados que já tem uma estrutura fixa. 

- Dados não estruturados
Alguns deles podem possuir formato determinado ou outros que não vão ter uma estrutua definida, possivelmente vão ter estrutur desconhecida. Nessa caso, vamos precisar prever como vamos fazer o cruzamento dessas informações para conseguir fazer uma análise. 

- Dados semi estruturados
São componentes que conseguem ser agrupados com base nas suas características. Um exemplo prático: digamos que temos uma conta no google drive, lá eu vou conseguir ter arquivo de texto, vou ter planilhas, vou ter power point, como também eu vou ter o google forms. Mesmo eu tenho um formato pré-estabelcido a informação ou o conteúdo que está dentro daquele ambiente especificamente, vai ter sua variação própria, sua característica, mas eu já vou ter uma expextativa de como eu vou tratar aquele conteúdo.

TIPOS DE ANÁLISE DE DADOS
- Descritiva
Vamos ter a pergunta: O que está acontecendo em meu negócio?
Na análise descritiva nós vamos ter o primeiro passo, o ponto de partida para qualquer outra análise de dados. É nela que vamos começar a fazer a coleta desses dados, vamos começar a organizar, a sumarizar esses dados que muitas vezes podem vir brutos. 

- Diagnóstico
Por que isso está acontecendo? 
Esse tipo de análise buscar entender o que está por trás de determinados eventos e padrões que podem ser identificados. Posso usar algoritmos de correlação.

- Preditiva
O que poderá acontecer?
Na análise preditiva é o momento que começamos a prever o futuro através da análise de dados, porque basicamente a análise preditiva é uma das formas mais poderosas de análise de dados. Nela vamos utilizar diversas técnicas de estatiticas para conseguir fazer previsões, estimativas. Vamos utilizar algoritmo de regressão e machine learning.

- Prescritiva
O que eu preciso fazer?
Na análise prescritiva eu consigo ver muito além da minha etapa de pensar estimativa. Eu posso obter recomendações para conseguir fazer a minha tomada de decisão. Vou usar deep learning e inteligência artificial.
