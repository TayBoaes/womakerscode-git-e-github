## Introdução à cultura agil

Intrutora: Jessica Agatha

### Cultura Ágil
 Abordagem ágil para entregar produtos melhores. Processos de entrega mais eficazes. 

 Valorização para se ter uma condução melhor: o indivíduo e interação, software em funcionamento, colaboração com o cliente e responder a mudanças. 

Os 12 princípios: 

Ambiente sustentável para trabalhar
Entregar qualidade
Simplicidade 
Auto-organização 

Agilidade não é entregar rápido! Não é sinônimo de velocidade. 

O que é ser ágil?
Agilidade é a capacidade de se adaptar rapidamente às mudanças gerando valor. Para tal, é essencial criar uma cultura de entregas incrementáveis de valor em ciclos cada vez mais curtos e de incansável melhoria continua.

### Produto Mínimo Viável (MVP)
- Mvp (minimum viable product) é a versão simplificada de um produto final em que o mínimo de funcionalidade é apresentando para testar o encaixe do produto no mercado; 
- Permite que uma ideia ou hipótese seja validada antes de desenvolver o produto.

Ciclo de melhoria continua: aprender, construir e medir. 
Lean inception: é o nome dado aos workshop colaborativo para alinhar um grupo de pessoas sobre o produto mínimo viável a ser construído. 

**Entrega Incremental e iterativa:**
Construir o rascunho, validar e ajustar a qualidade conforme os feedbacks. 

É preciso inovar com foco em resolver as dores dos clientes! Temos que estar atentas aos problemas e resovê-las!

### O que é Scrum

Os pilares e valores do Scrum são: transparência, Inspeção, Adapatação  

**Sobre o Scrum**
- Fácil de entender, mas difícil de dominar;
- Sprints de 1 a 4 semanas; 
- É um framework, não um método.
- A cada entrega, uma inspeção (retrospectiva) e adaptação para o próximo ciclo (sprint).

### DoR e Dod 

DOR - Definition of ready
DOD - Definition of done 

São acordos. 

**Métrica de eficácia e feiciência**
O que eu tenho que fazer e como eu vou fazer. 

Métricas devem ser usadas para evoluir o processo e não para gerar cobrança e comparações destrutivas. As métricas são um convite para conversa.
- Planejamento com base em dados reais e não em sentimento e achismos
- Entendimento sobre quais são os pontos de melhoria, para então priorização da atuação sobre eles
- Análise sobre a forma como se trabalha 

Uma das métricas importantes é o "Lead Time" que é o tempo considerado para o desenvolvimento do trabalho. 

**O que é um time ágil?**
Formar um time ágil vai além de juntar pessoas. É preciso entender claramente o propósito deste time antes de criá-lo, uma vez que ele precisa conter os conhecimentos, habilidades e autonomia necessária para fazer a entrega de ponta a ponta.
